```python
from tkinter import *
from tkinter.ttk import *
from tkinter import ttk
import tkinter as tk
import time
from tkinter import filedialog
import os
import csv
import random
from unidecode import unidecode
from tkinter import messagebox


def loadFile_dict():
    global objfile_dict
    file = filedialog.askopenfilenames()
    objfile_dict = file
    LoadFile.configure(text = objfile_dict) #changes the name of button Encrypt
def func(lvar):
    global window
    global result
    global txt
    global ans
    if unidecode(txt.get()).lower() == unidecode(ans).lower() :
        result.configure(text ="Right!",foreground = "green")
    else:
        result.configure(text ="Wrong! \nThe answer is: "+ ans ,foreground = "red")
    result.update()
    time.sleep(2)
    window.destroy()
def start():
    global root
    global window
    global txt
    global result
    global objfile_dict
    global ans
    
    timer = int(combo.get())
    if os.path.splitext(objfile_dict[0])[-1].lower() != ".csv" and os.path.splitext(objfile_dict[0])[-1].lower() != ".txt":
        messagebox.showwarning('Error', 'Dictionary file must be a csv or txt file!')
        return
    root.destroy()
    options = []
    with open(objfile_dict[0]) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        for row in csv_reader:
            options.append(row)
    
    while True:
        rng = random.randint(0,1)
        test = options[random.randint(0, len(options) -1)]
        ans = test[rng]
        time.sleep(timer)
        window = Tk()
        lbl = Label(window, text= test[rng - 1])
        lbl.pack()
        txt = Entry(window,width=60)
        txt.pack()
        txt.focus()
        result = Label(window, text = " Translate")
        result.pack()
        window.bind('<Return>', func)
        window.mainloop()
    

root = Tk()
combo = Combobox(root)
lbl = Label(root, text="Dictionary List:")
lbl.pack()
LoadFile = Button(root,text = "Load file", command = loadFile_dict)
LoadFile.pack()
lbl2 = Label(root, text="Time intervals:")
lbl2.pack()
combo['values']= (0, 10, 20, 30, 45, 60, 120, 240, 500)
combo.current(0) #set the selected item
combo.pack()
Button(root, text="Start", command=start).pack()
root.mainloop()
```
